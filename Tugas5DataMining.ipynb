```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
```

```python
df = pd.read_csv('DataTugas.csv')
df
```

```python
Tidak = df[df["Rekomendasi"] == "Tidak"]
Ya = df[df["Rekomendasi"] == "Ya"]

data = (Tidak, Ya)
colors = ("red","green")
groups = ("Tidak", "Ya")

for data, color, group in zip(data, colors, groups):
    x, y = data["Kehadiran"], data["Lingkungan"]
    plt.scatter(x, y, c=color, label=group)

plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
X = df.iloc[:, :4]
y = df.iloc[:, 4]
print(X)
print(y)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X,y)
```

```python
#Ucha kehadiran rajin, lingkungan tidak peduli, kerjasama tidak mampu, prakarsa kurang inisiatif
data_test = {
    'Kehadiran' : [4],
    'Lingkungan' : [1.8],
    'Kerjasama' : [1.5],
    'Prakarsa' : [3]
}

test = pd.DataFrame(data_test)
knn.predict(test)
```
